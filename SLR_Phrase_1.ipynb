```python
# Import the pandas libraray 
import pandas as pd 
```

```python
# Load the "Salary_Data.csv" file using Pandas and store the file in the variable "dataset"
dataset=pd.read_csv("Salary_Data.csv")
```

```python
# Print The Dataset
dataset
```

```python
# Load the Yearsexperience and Store it in Independent
Independent  = dataset[["YearsExperience"]]
```

```python
# Print the variable
Independent
```

```python
# Load the Salary and Store it in Dependent
Dependent  = dataset[["Salary"]]
```

```python
#Print the storee d data
Dependent
```

```python
# we'll import the sci-kit learn from sci-kit learn we'll import the function train_test_split 
from sklearn.model_selection import train_test_split
```

```python
# Split the Independent using the variable named X_train, X_test
# split the Dependent using the variable named y_train, y_test
X_train, X_test, y_train, y_test = train_test_split(Independent, Dependent, test_size=0.20, random_state=0)
```

```python
# We'll import the algorithm named LinearRegression from the Sci-kit learn
from sklearn.linear_model import LinearRegression
```

```python
# Store the LinearRegression into variable named regression
regression = LinearRegression()
```

```python
# We should check the fit 
regression.fit(X_train, y_train)
```

```python
# Store the data into weight
weight = regression.coef_
```

```python
# Store the weight data and print it
weight
```

```python
# Store the data into bais
bais = regression.intercept_
```

```python
# Store the bais data and print it
bais
```

Evaluation Matrix (or) Validation Parameter

```python
# Create the variable named "y_pred" and saved the X_test data in the varibale
y_pred = regression.predict(X_test)
```

```python
#Evaluationm parameters import r2_score from the metrics in sklearn
from sklearn.metrics import r2_score  
```

```python
#To find the accuracy score in the variable "r_score"
r_score = r2_score(y_test,y_pred)
```

```python
#Print the accuray score it will near to 1 it is good one
r_score 
```
